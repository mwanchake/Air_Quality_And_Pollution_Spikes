```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pickle

# 1️⃣ Load dataset
data = pd.read_csv("airQuality.csv")  

# 2️⃣ Select features (X) and target (y)
X = data[['pm2.5', 'pm10', 'temperature']]
y = data['predicted_aqi']

# 3️⃣ Split dataset (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# 4️⃣ Create and train model
model = LinearRegression()
model.fit(X_train, y_train)

# 5️⃣ Save the trained model
with open("aqi_model.pkl", "wb") as file:
    pickle.dump(model, file)

print("Model saved successfully as aqi_model.pkl")
```
